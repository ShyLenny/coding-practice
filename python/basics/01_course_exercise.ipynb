```python
def findDog(st):
    return 'dog' in st.lower().split()

findDog('Is there a dog here?')
```

```python
def countDog(st):
    count = 0
    for word in st.lower().split():
        if word == 'dog':
            count += 1
    return count

countDog('This dog runs faster than the other dog dude!')
```
