```python
# array range (start, stop, step)
a= np.arange(0, 10, 2)
print(a)
```

```python
# array of zeros (rows, columns)
z= np.zeros((2, 3))
print(z)
```

```python
# array of ones (rows, columns)
o= np.ones((3, 4))
print(o)
```

```python
# array of evenly spaced numbers over a specified interval (start, stop, number of samples)
s= np.linspace(0, 10, 5)
print(s)
```

```python
# identity matrix (size)
i= np.eye(3)
print(i)
```

```python
# array of random numbers (rows, columns)
r= np.random.rand(5,2)
print(r)
```

```python
# from standard normal distribution or from a gaussion distribution
n= np.random.randn(3, 4)
print(n)
```

```python
# array of random integers (low, high, size)
m= np.random.randint(1, 10, (3, 4))
print(m)
```

```python
# reshape an array
m= m.reshape(5, 1)
print(m)
```

```python
# maximum and minimum values
max_val = np.max(m)
print("Maximum value:", max_val)
min_val = np.min(m)
print("Minimum value:", min_val)
```

```python
# index of maximum and minimum values
max_index = np.argmax(m)
print("Index of maximum value:", max_index)
min_index = np.argmin(m)
print("Index of minimum value:", min_index)
```

```python
# shape of an array , no parenthesis needed
shape = m.shape
print("Shape of the array:", shape)
```

```python
ar= np.arange(0, 11)
print(ar)
print(ar[8])
print(ar[1:5])
print(ar[5:])
```

```python
#NUMPY ARRAYS DIFFER FROM array lists because of their ability to broadcast.
# This allows for element-wise operations on arrays of different shapes.
#  This makes numpy arrays more efficient for numerical computations compared to standard Python lists.

slice= ar[0:6]
print(slice)
```

```python
slice[:]= 99
print(slice)
print (ar)
```

```python
ar_copy= ar.copy()
ar_copy[:]= 100
print(ar_copy)
```
